````python
import random

# 4-letter words with distinct letters
WORDS = [
    'bake', 'fire', 'gold', 'milk', 'dawn', 'wind', 'fork', 'lamp',
    'ship', 'tide', 'rock', 'move', 'hunt', 'sand', 'pear', 'wolf',
    'drop', 'beam', 'leaf', 'plot', 'play'
]

def valid(word):
    return len(word) == 4 and len(set(word)) == 4 and word.isalpha()

def feedback(guess, target):
    stars = sum(g == t for g, t in zip(guess, target))
    stars_empty = sum((g in target) for g in guess) - stars
    return stars, stars_empty

def filter_words(words, guess, stars, stars_empty):
    result = []
    for w in words:
        s, e = feedback(guess, w)
        if s == stars and e == stars_empty:
            result.append(w)
    return result

print("Think of a 4-letter English word ğŸ’­ (letters not repeated)")
print("I'll try to guess it. You can also try to guess mine! Let's start ğŸ‘±ğŸ†šğŸ¤–")

ai_word = random.choice(WORDS)
possible = WORDS.copy()

while True:
    # ğŸ‘± Human turn
    player_guess = input("\nYour ğŸ‘± guess: ").lower().strip()
    if not valid(player_guess):
        print("Please enter a valid 4-letter word with unique letters.")
        continue

    s, e = feedback(player_guess, ai_word)
    print(f"Feedback: {s}â­ and {e}â˜†")

    if s == 4:
        print("You win! ğŸ‰ Smart human!")
        break

    # ğŸ¤– AI turn
    if not possible:
        print("Hmm, something's wrong â€” no possible words left ğŸ˜…")
        break

    ai_guess = random.choice(possible)
    print(f"\nAI ğŸ¤– guesses: {ai_guess}")

    try:
        s_ai = int(input("How many â­ (correct letters, right place): "))
        e_ai = int(input("How many â˜† (correct letters, wrong place): "))
    except ValueError:
        print("Invalid input. Please enter numbers.")
        continue

    if s_ai == 4:
        print(f"AI ğŸ¤– wins! I guessed your word {ai_guess}! ğŸ’ª")
        break

    possible = filter_words(possible, ai_guess, s_ai, e_ai)
    print(f"AI ğŸ¤– now has {len(possible)} possible words left to consider.")

print("Game over. Thanks for playing!")
````
